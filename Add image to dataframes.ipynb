```python
import pandas as pd
from IPython.core.display import HTML
import numpy as np
import requests
import re
from bs4 import BeautifulSoup

pd.set_option('display.max_colwidth', -1)
```

### JTI

```python
#load dataframe
jti = pd.read_excel("Japan Tobacco Inc.xlsx")
#create blank column to store image src
jti['Image'] = ''
#clean up goods/service column
for i in jti['GOODS / SERVICES']:
    if '34' in i:
        jti = jti.replace(i, '34')
```

```python
#load txt file for query
file = BeautifulSoup(open("html_raw.txt", encoding = 'utf8').read(), "lxml")
```

```python
for number in jti['APP. NUMBER']:
    for table in file.find_all('table'):
        #if the app number is present in the table:
        if number in str(table):
            #append src tag of the image to the current dataframe
            jti['Image'][jti.loc[jti['APP. NUMBER']==number].index[0]] = table.find_all('img')[0]
```

```python
HTML(jti.to_html(escape=False ,formatters=dict(image=jti['Image'])))
```
