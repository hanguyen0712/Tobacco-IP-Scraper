```python
import pandas as pd
import numpy as np
import requests
import csv
import os
import xlsxwriter
import openpyxl
from openpyxl import load_workbook
from openpyxl import Workbook
from openpyxl.drawing.image import Image
from openpyxl.utils import coordinate_from_string
import PIL
from PIL import Image
```

```python
#load data
df = pd.read_excel("Everything.xlsx", 0)
df.head()
```

```python
len(df)
```

```python
#save path
save_path = 'C:/Users/admin/Desktop/projects/Tobacco-IP-Scraper/img'

#retrieve images:
for i in range(0, len(df)):
    filename = "{}.{}".format(df['APP. NUMBER'][i], df['Image'][i][-3:])
    completename = os.path.join(save_path, filename)  
    url = df['Image'][i]
    result = requests.get(url, stream = True)
    if result.status_code == 200:
        image = result.raw.read()
        open(completename, 'wb').write(image)
        print(i)
```

```python
#load workbook
wb = load_workbook("Data Table Filled_030719.xlsx")

#append image to worksheet
for sheet in range(0, 3):
    #open pd dataframe for each sheet - nvm the dataframe name
    jpi = pd.read_excel('Data Table Filled_030719.xlsx', sheet)

    #open workbook to insert picture
    
    ws = wb.worksheets[sheet]
    for i in range(0, len(jpi['APP. NUMBER'])):
        img_loc = r'C:/Users/admin/Desktop/projects/Tobacco-IP-Scraper/img/{}.{}'.format(jpi['APP. NUMBER'][i], jpi['Image link'][i][-3:])
        img = Image.open(img_loc)
        wsize = int((float(img.size[0])*float(75/float(img.size[1]))))
        img = img.resize((wsize, 75), Image.ANTIALIAS)
        img.save(img_loc, 'jpeg')
        img = openpyxl.drawing.image.Image(img_loc)
        ws.add_image(img, 'L{}'.format(i+2))
        print(i)

wb.save("Data Table Filled_030719.xlsx")
```
